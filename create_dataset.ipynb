```python
def increase_brightness(img):
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    
    value = np.random.randint(30, size=1)[0]
    
    vValue = hsv[...,2]
    hsv[...,2] = np.where((255-vValue)<value,255,vValue+value)
    
    value = np.random.randint(35, size=1)[0]/100+0.65
    
    hsv[...,2] = hsv[...,2]*value
    
    img = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)
    return img
def transform(r, g, b):
    r, g, b = (r, g, b)
    avg_v = r * 0.2126 + g * 0.7152 + b * 0.0722
    value = (np.random.randint(100, size=3)+0.1)/100
    r += (r - avg_v) * value[0]
    g += (g - avg_v) * value[1]
    b += (b - avg_v) * value[2]
    return r, g, b

def blur_rand_place(img):
    
    numb = np.random.randint(15, size=1)[0]+10
    
    for i in range(numb):
    
        h = np.random.randint(img.shape[0]//6, size=1)[0]+10
        y = np.random.randint(img.shape[0], size=1)[0]
        
        w = np.random.randint(img.shape[1]//6, size=1)[0]+10
        x = np.random.randint(img.shape[1], size=1)[0]

        y0 = y
        y1 = min(y+h,img.shape[0])
        x0 = x
        x1 = min(x+w, img.shape[1])

        img[y0:y1, x0:x1] = cv2.blur(img[y0:y1, x0:x1] ,(5,5))
        
    return img
```

```python
def cluster_points(descriptors):
    descriptors = StandardScaler().fit_transform(descriptors)
    cluster = MeanShift(bandwidth=0.4)
    #cluster = OPTICS(min_samples=3)
    return cluster.fit_predict(descriptors)

def compute_mask(image, points):
    hull = ConvexHull(points)
    h, w, _ = image.shape
    mask = np.zeros((h, w))
    cv2.fillConvexPoly(mask, hull.points[hull.vertices].astype('int'), 255)
    return mask
def transform_image(image0, image1, mask, points0, points1):
    h, _ = cv2.findHomography(points1, points0, cv2.RANSAC)
    height, width, c = image0.shape
    if h is not None:
        
        transformed = cv2.warpPerspective(image1, h, (width, height))
        mask = cv2.warpPerspective(mask, h, (width, height))
        return transformed, mask
    return None, None
def apply_mask(image, transformed, mask, inplace=False):
    if not inplace:
        image = image.copy()     
    image[mask.astype(bool)] = transformed[mask.astype(bool)]
    return image
def pipeline(points0, points1, image0, image1):
    n=0
    #print('Doing clustering...')
    clusters = cluster_points(points1)
    #print(np.unique(clusters))
    image0 = image0.copy()
    #print('Processing clusters...')
    result_mask = np.zeros((image0.shape[0],image0.shape[1]))
    for c in np.unique(clusters):
        if np.sum(clusters == c) > 2:
            
            mask = compute_mask(image1, points1[clusters == c])
            transformed, mask = transform_image(image0, image1, mask, points0[clusters == c], points1[clusters == c])
            
            if transformed is not None:
                result_mask+=mask
                image0 = apply_mask(image0, transformed, mask, inplace=True)
                n=1
            
 
    return image0, result_mask, n
```

```python
def create_proc(img_path1):
    
    img_path2 = in_dir2+img_path1.split('/')[-1]

    image0, inp0, scales0 = read_image(img_path1, device, RESIZE, ROT0, RESIZE_FLOAT)
    image1, inp1, scales1 = read_image(img_path2, device, RESIZE, ROT1, RESIZE_FLOAT)

    mask_path1 = in_dir_m1 + img_path1.split('/')[-1]
    mask_path2 = in_dir_m2 + img_path1.split('/')[-1]
    
    mask1 = cv2.imread(mask_path1, cv2.IMREAD_GRAYSCALE)
    mask2 = cv2.imread(mask_path2, cv2.IMREAD_GRAYSCALE)
    

    inp0_new = torch.tensor((cv2.resize(mask1,(image0.shape[1],image0.shape[0]))/255.).astype(np.uint8)) * inp0
    inp1_new = torch.tensor((cv2.resize(mask2,(image1.shape[1],image1.shape[0]))/255.).astype(np.uint8)) * inp1
    
    pred = matching({'image0': inp0_new, 'image1': inp1_new})
    pred = {k: v[0].cpu().numpy() for k, v in pred.items()}
    kpts0, kpts1 = pred['keypoints0'], pred['keypoints1']
    matches, conf = pred['matches0'], pred['matching_scores0']
    out_matches = {'keypoints0': kpts0, 'keypoints1': kpts1,
                   'matches': matches, 'match_confidence': conf}
    
    valid = matches > -1
    mkpts0 = kpts0[valid]
    mkpts1 = kpts1[matches[valid]]
    
    #fig, ax = plt.subplots(figsize=(45, 40))
    #ax.imshow(image0, cmap='Greys_r')
    #ax.scatter(mkpts0[:, 0], mkpts0[:, 1], s=10, c='r')
    
    image0 = cv2.imread(img_path1)
    input_h,input_w = image0.shape[1],image0.shape[0]
    image1 = cv2.imread(img_path2)[..., [2, 1, 0]]
    
    #image0 = increase_brightness(image0)
    image0 = cv2.cvtColor(image0, cv2.COLOR_BGR2RGB)
    image0 = Image.fromarray(image0)
    values = np.random.randint(63, size=3)+2
    lut = Color3DLUT.generate(values, transform)
    image0 = image0.filter(lut)

    # For reversing the operation:
    image0 = np.asarray(image0)
    image0 = cv2.resize(image0,(input_h//3,input_w//3),interpolation=cv2.INTER_CUBIC )
    image0 = cv2.resize(image0,(input_h,input_w))
    
    image0 = cv2.cvtColor(image0, cv2.COLOR_RGB2BGR)[..., [2, 1, 0]]
 
    mask1 = (cv2.resize(mask1,(image0.shape[1],image0.shape[0]))/255.).astype(np.uint8)
    mask2 = (cv2.resize(mask2,(image1.shape[1],image1.shape[0]))/255.).astype(np.uint8)
    
    mkpts1_new = []
    mkpts0_new = []
    for indx0,indx1 in zip(mkpts0*scales0,mkpts1*scales1):
        if (mask1[int(indx0[1]),int(indx0[0])]==1 and mask2[int(indx1[1]),int(indx1[0])]==1):
            mkpts0_new.append(indx0)
            mkpts1_new.append(indx1)

    if len(mkpts0_new)<4:
        mkpts0*=scales0
        mkpts1*=scales1
        
    else:
        mkpts0 = np.array(mkpts0_new)
        mkpts1 = np.array(mkpts1_new)

    return mkpts0, mkpts1, image0, image1
```

```python
in_dir2 = '/home/ekaterina/Work/Dataset/outdoor_dataset/source/'
in_dir1 = '/home/ekaterina/Work/Dataset/outdoor_dataset/target/'
in_dir_m2 = '/home/ekaterina/Work/Dataset/outdoor_dataset/source_mask/'
in_dir_m1 = '/home/ekaterina/Work/Dataset/outdoor_dataset/target_mask/'

out_dir_m1 = '/home/ekaterina/Work/Dataset/outdoor_dataset/source_transforemed/'
out_dir_m2 = '/home/ekaterina/Work/Dataset/outdoor_dataset/target_mask_transformed/'

numb = 65+30+46+32+25+33+35

source_images1 = sorted(glob(join(in_dir1, '*.jpg')))[numb:]

for img_path1 in tqdm(source_images1):

    mkpts0, mkpts1, image0, image1 = create_proc(img_path1)
    
    transformed_image0, mask_transform, n = pipeline(mkpts0, mkpts1, image0, image1)
    
    #transformed_image0 = cv2.cvtColor(transformed_image0, cv2.COLOR_BGR2RGB)

    if n==1:
        imageio.imwrite(out_dir_m1+img_path1.split('/')[-1],transformed_image0)
        imageio.imwrite(out_dir_m2+img_path1.split('/')[-1],mask_transform.astype(np.uint8))
    
    gc.collect()
```
